```python
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from scipy import ndimage, signal
import math


%matplotlib inline


image_bank = {}
```

```python
def get_image(filename='img/money1.png'):
    if filename in image_bank.keys():
        return image_bank[filename]
    
    img = cv2.imread(filename)
    image_bank[filename] = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return image_bank[filename]


 
def plot_image(images, title=None, labels=None, size=(13, 4)):
    if not isinstance(images, list):
        images = [images]
 
    if not isinstance(labels, list):
        labels = []
 
    fig, axs = plt.subplots(1, len(images), figsize=size,  dpi=150)
 
    if title is not None:
        fig.suptitle(title, fontsize=16)
    need_titles = len(labels) == len(images)
    for i, image in enumerate(images):
        ax = axs[i] if len(images) > 1 else axs
 
        if need_titles:
            ax.set_title(labels[i])
        ax.imshow(image, cmap='gray')
        ax.axis('off')
     
    
def CLAHE(img):
    #-----Converting image to LAB Color model----------------------------------- 
    lab= cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

    #-----Splitting the LAB image to different channels-------------------------
    l, a, b = cv2.split(lab)
#     cv2.imshow('l_channel', l)
#     cv2.imshow('a_channel', a)
#     cv2.imshow('b_channel', b)

    #-----Applying CLAHE to L-channel-------------------------------------------
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl = clahe.apply(l)

    #-----Merge the CLAHE enhanced L-channel with the a and b channel-----------
    limg = cv2.merge((cl,a,b))

    #-----Converting image from LAB Color model to RGB model--------------------
    final = cv2.cvtColor(limg, cv2.COLOR_LAB2RGB)
    return final
```

```python
original_image = np.abs(np.power(get_image().astype(np.uint16), 1.7))

# original_image = CLAHE(cv2.imread('img/money1.png'))
# original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

original_image = (original_image / original_image.max() * 255).astype(np.uint8)
plot_image(original_image, title='Original image')
```

```python
orig_adapt_mean = cv2.adaptiveThreshold(original_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 451, 7)

plot_image(orig_adapt_mean, labels=['Gaussian binarization'])
```

```python
def ring(r):
    d = 2 * r + 1
    rx, ry = d / 2, d / 2
    x, y = np.indices((d, d))
    distance = np.hypot(rx - x, ry - y)
    mask = np.abs(distance - r) < 0.5
    return mask.astype('uint8')
```

```python
def erode(image, kernel, iterations=1):
    return cv2.erode(image, kernel, iterations)


def ellipse(size):
    return cv2.getStructuringElement(cv2.MORPH_ELLIPSE, size)


def dilate(image, kernel, iterations=1):
    return cv2.dilate(image, kernel, iterations)


def closure(image, kernel):
    return cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)


# plot_image(erode(dilate(np.abs(orig_adapt_mean - 1), ring(60)), ellipse((13, 13))))


dilated = np.abs(dilate(np.abs(orig_adapt_mean - 1), ring(60)) - 255)
filtered = erode(dilate(dilated, ring(3)), ellipse((25, 25)))
ready = dilate(filtered, ellipse((120, 120)))
mask = np.abs(dilate(erode(filtered, ellipse((100, 100))), ellipse((530, 530))) - 255)
ready *= mask
```

```python
big_coins = dilate(erode(ready, ring(100)), ellipse((120, 120)))
small_coins = dilate(erode(ready, ring(60)) * (np.abs(big_coins - 255)), ellipse((120, 120)))
```

```python
def get_centroids(image):
    from skimage.measure import label, regionprops
    
    image = np.array(image)
    label_img = label(image)
    regions = regionprops(label_img)
    return np.array([region.centroid for region in regions])


print(f'Found {len(get_centroids(big_coins))} big coins')
plot_image(big_coins)
```

```python
print(f'Found {len(get_centroids(small_coins))} small coins')
plot_image(small_coins)
```

```python
img_color = (ready * get_image()).copy()
img_gray = cv2.GaussianBlur(img_color, (7, 7), 0)

circles = cv2.HoughCircles(img_gray, cv2.HOUGH_GRADIENT, 60, minDist=200,
                     param1=90, param2=100, minRadius=115, maxRadius=125)

if circles is not None:
    img_color = cv2.cvtColor(img_color, cv2.COLOR_GRAY2BGR)
    for i in circles[0, :]:
        # draw the outer circle
        cv2.circle(img_color, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # draw the center of the circle
        cv2.circle(img_color, (i[0], i[1]), 2, (0, 0, 255), 3)
        
plot_image(img_color)
```

```python
# Rhys Dunn - 2015
# Learning image vision/OpenCV


# import libraries
import numpy as np
import cv2


# load image
img = cv2.imread("img/money1.png")


# prep image - blur and convert to grey scale
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.medianBlur(grey, 39)


# canny edge detector
outline = cv2.Canny(blurred, 40, 150)


# find the contours
(cnts, _) = cv2.findContours(outline, cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_SIMPLE)


# draw contours: -1 will draw all contours
cv2.drawContours(img, cnts, -1, (0, 255, 0), 2)

plot_image(img)

# Print how many coins we found
print("I found %i coins" % len(cnts))
```
