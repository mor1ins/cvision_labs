```python
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

%matplotlib inline


image_bank = {}
```

```python
def get_image(filename='img/gears1.png'):
    if filename in image_bank.keys():
        return image_bank[filename]
    
    img = cv2.imread(filename, 0)
    image_bank[filename] = img
    return image_bank[filename]


def plot_image(image, size=(3, 3)):
    fig, ax = plt.subplots(figsize=size,  dpi=300)
    ax.imshow(image, cmap='gray')
    ax.axis('off')
    plt.show()

    
    
def ring(r):
    d = 2 * r + 1
    rx, ry = d / 2, d / 2
    x, y = np.indices((d, d))
    return (np.abs(np.hypot(rx - x, ry - y) - r) < 0.5).astype('uint8')
    

plot_image(get_image())
```

```python
def erode(image, kernel, iterations=1):
    return cv2.erode(image, kernel, iterations)


kernel = ring(70)
step1 = erode(get_image(), kernel)
plot_image(step1, size=(3, 3))
```

```python
def dilate(image, kernel, iterations=1):
    return cv2.dilate(image, kernel, iterations)


kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (150, 150))
step2 = dilate(step1, kernel)
plot_image(step2)
```

```python
step3 = get_image().copy()
step3[step2 > 100] = step1.max()
plot_image(step3)
```

```python
kernel = ring(1)
step4 = cv2.morphologyEx(step3, cv2.MORPH_CLOSE, kernel)

kernel = ring(2)
step4 = dilate(step4, kernel)

kernel = ring(10)
step4_opening = cv2.morphologyEx(step3, cv2.MORPH_OPEN, kernel)

kernel = ring(6)
step4_opening = dilate(step4_opening, kernel)

step4[step4_opening > 100] = step4.min()

step4[step4 > 50] = 255
# plot_image(dilate(step4, kernel))
plot_image(step4)
```

```python
gear_body_size = 270
gear_body = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (gear_body_size, gear_body_size))
closure = cv2.morphologyEx(step3, cv2.MORPH_OPEN, gear_body)

sampling_ring_spacer_size = 30
sampling_ring_spacer = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,
(sampling_ring_spacer_size, sampling_ring_spacer_size))
dilatation_bw = cv2.dilate(closure, sampling_ring_spacer)

res = cv2.subtract(dilatation_bw, closure)

plot_image(res)
```

```python
size = 27
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (size, size))
step5 = dilate(step4, kernel)
plot_image(step5)
```

```python
size = 15
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (size, size))
step6 = cv2.subtract(erode(res, kernel), step5)
step6[step6 > 50] = 255

size = 40
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (size, size))
plot_image(dilate(step6, kernel))
```

```python
from PIL import Image



images = [erode(get_image(), ring(i)) for i in range(120)]    
images_new = np.vectorize(np.vectorize(lambda x: (x, x, x)))(images)


images_new[0].save('img/result.gif',
               save_all=True, append_images=images_new[1:], optimize=False, duration=120, loop=0)
```
